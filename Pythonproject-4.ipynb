```python
#24 Hour Daily News Flash
#Created by Maryam
#I made this project to show how news changes every hour
import datetime
import time
from IPython.display import clear_output

# 24 headlines - one for each hour of the day
headlines = [
    "Google releases AI tool that helps students research faster",
    "Srinagar University opens applications for data science program",
    "Malaysia ranked among the top countries for international students",
    "Python remains the most used language in data science",
    "Kashmir valley gets better internet in rural areas",
    "Malaysian universities offering more scholarships this year",
    "Data science jobs increasing rapidly across Southeast Asia",
    "Srinagar students join national coding competition",
    "Microsoft launches free online AI course for beginners",
    "Study shows data skills becoming important in every field",
    "Kashmir girl builds model to monitor pollution in Dal Lake",
    "Kuala Lumpur hosts biggest tech conference in Southeast Asia",
    "Python now being taught in schools across South Asia",
    "Malaysia investing in AI research for national development",
    "Srinagar student wins award for data visualisation project",
    "Demand for data scientists expected to grow 35 per cent by 2030",
    "Tech companies in Malaysia hiring more fresh graduates",
    "Kashmir University signs exchange program with Malaysian college",
    "OpenAI releases tools to help students learn machine learning",
    "Data science improving healthcare systems across Asia",
    "Srinagar coding club teaches Python to 200 students this year",
    "Malaysia introduces digital skills program for students",
    "Google certification now accepted by top Asian employers",
    "Tech industry in Asia continues to grow at record pace",
]

def get_headline_for_hour(hour):
    return headlines[hour]

def show_headline(headline, hour):
    
    # Get current time
    now = datetime.datetime.now()
    
    # Using f-string instead of strftime
    hour_now   = now.hour
    minute_now = now.minute
    day_now    = now.day
    month_now  = now.month
    year_now   = now.year
    
    time_now = f"{hour_now}:{minute_now}  |  {day_now}/{month_now}/{year_now}"

    # Pick morning/afternoon/evening/night
    if hour < 12:
        period = "Morning News"
    elif hour < 17:
        period = "Afternoon News"
    elif hour < 21:
        period = "Evening News"
    else:
        period = "Night News"

    print("=" * 62)
    print("       Daily News Flash - Maryam's Project")
    print("=" * 62)
    print("  Time:  " + time_now)
    print("  " + period)
    print("=" * 62)
    print()
    print("  " + headline)
    print()
    print("=" * 62)
    print("  Hour " + str(hour) + " of 24")
    print("=" * 62)

# Run through all 24 hours
print("Loading headlines..")
time.sleep(2)

for hour in range(24):
    clear_output(wait=True)
    headline = get_headline_for_hour(hour)
    show_headline(headline, hour)
    time.sleep(2)

clear_output(wait=True)
print("=" * 62)
print("  All done - 24 headlines for 24 hours")
print("=" * 62)
```
